```python
#Unsupervised Classification -  Using KMeans

#Soil
#Crop residue
#Sugarcane Crop
#Crop Weed

import numpy as np
import cv2
from matplotlib import pyplot as plt
import matplotlib.image as mpimg

from sklearn.cluster import KMeans
```

```python
w=800
h=600
#K Groups for clustering
K = 4
```

```python
#Read and Resize Image
img = cv2.imread('images/slice.jpg')
resized = cv2.resize(img,(w,h))

#Print Original Image
plt.imshow(resized)
plt.show()
```

```python
#Original Image Histogram
plt.hist(img.ravel(),256,[0,256]); 
plt.show()

#Resized Image Histogram
#plt.hist(resized.ravel(),256,[0,256]); 
#plt.show()
```

```python
Z = resized.reshape((-1,3))
Z = np.float32(Z)
```

```python
#Clustering 
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)

# Now separate the data, Note the flatten()
A = Z[label.ravel()==0]
B = Z[label.ravel()==1]
C = Z[label.ravel()==2]
D = Z[label.ravel()==3]

# Plot the data
plt.scatter(A[:,0],A[:,1],c = 'r')
plt.scatter(B[:,0],B[:,1],c = 'g')
plt.scatter(C[:,0],C[:,1],c = 'b')
plt.scatter(D[:,0],D[:,1],c = 'purple')

plt.scatter(center[:,0],center[:,1],s = 80,c = 'y', marker = 's')
#plt.xlabel('Height'),plt.ylabel('Weight')
plt.show()
```

```python
center = np.uint8(center)
result = center[label.flatten()]
result2 = result.reshape((resized.shape))

print(center)
```

```python
#cv2.imshow('unsupervised kmeans K:3',res2)
plt.imshow(result2)
plt.show()
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
plt.hist(result2.ravel(),256,[0,256]); 
plt.show()
```

```python
#clase 1 (r)
result2[np.where((result2 == center[0]).all(axis = 2))] = [255,0,0]
#clase 2 (g)
result2[np.where((result2 == center[1]).all(axis = 2))] = [0,255,0]
#clase 3 (b)
result2[np.where((result2 == center[2]).all(axis = 2))] = [0,0,255]
#clase 4 (purple)
result2[np.where((result2 == center[3]).all(axis = 2))] = [199,149,238]

plt.imshow(result2)
plt.show()
```

```python
#Another way to compute Kmeans 
#Using Kmeans from scikit
model = KMeans(n_clusters=K) #Set the model to have 20 clusters
model = model.fit( Z )

mc=model.cluster_centers_
ml=model.labels_

# Now separate the data, Note the flatten()
A = Z[ml.ravel()==0]
B = Z[ml.ravel()==1]
C = Z[ml.ravel()==2]
D = Z[ml.ravel()==3]

# Plot the data
plt.scatter(A[:,0],A[:,1],c = 'r')
plt.scatter(B[:,0],B[:,1],c = 'g')
plt.scatter(C[:,0],C[:,1],c = 'b')
plt.scatter(D[:,0],D[:,1],c = 'purple')

plt.scatter(mc[:,0],mc[:,1],s = 80,c = 'y', marker = 's')
plt.show()

mc = np.uint8(mc)
result4 = mc[ml.flatten()]
result4 = result4.reshape((resized.shape))

#clase 1 (r)
result4[np.where((result4 == mc[0]).all(axis = 2))] = [255,0,0]
#clase 2 (g)
result4[np.where((result4 == mc[1]).all(axis = 2))] = [0,255,0]
#clase 3 (b)
result4[np.where((result4 == mc[2]).all(axis = 2))] = [0,0,255]
#clase 4 (purple)
result4[np.where((result4 == mc[3]).all(axis = 2))] = [199,149,238]

plt.imshow(result4)
plt.show()
```
